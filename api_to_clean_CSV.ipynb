```python
titles = []
metascore = []
imdb_rating = []
imdb_votes = []
rated = []
runtime = []
released = []
genre = []
box_office = []
year = []

imdb_source = []
imdb_score = []

rotten_source = [] # [DL] not used yet. depends on if we want to include the rotten tomatoe info
rotten_score = []

metacrit_source = [] # [DL] not used yet. depends on if we want to include the metacritic info
metacrit_score = []

notfound = 0
found = 0
total = len(df)
j = 0
```

```python
for film in df['Title']:
    try:
        response = omdb.get(title={film}) 
        titles.append(response["title"])
    except:
        titles.append('null')
    try:
        box_office.append(response["box_office"])
    except:
        box_office.append('null')
    try:
        metascore.append(response["metascore"])
    except:
        metascore.append('null')
    try:
        imdb_rating.append(response["imdb_rating"])
    except:
        imdb_rating.append('null')
    try:
        imdb_votes.append(response["imdb_votes"])
    except:
        imdb_votes.append('null')
    try:
        rated.append(response["rated"])
    except:
        rated.append('null')
    try:
        genre.append(response["genre"])
    except:
        genre.append('null')
    try:
        runtime.append(response["runtime"])
    except:
        runtime.append('null')
    try:
        released.append(response["released"])
    except: 
        released.append('null')
    try:
        year.append(response["year"])
    except:
        year.append('null')
    try: 
        imdb_source.append(response['ratings'][0]['source']) #IMDB source 
    except:
        imdb_source.append('null')
    try:
        imdb_score.append(response['ratings'][0]['value']) #IMDB score
    except:
        imdb_score.append('null')
    try: 
        rotten_source.append(response['ratings'][1]['source']) #rotten tomatoes source
    except:
        rotten_source.append('null')
    try:
        rotten_score.append(response['ratings'][1]['value']) #rotten tomatoes score
    except:
        rotten_score.append('null') 
    try:
        metacrit_source.append(response['ratings'][2]['source']) #metacritic source
    except:
        metacrit_source.append('null') #metacritic source
    try:
        metacrit_score.append(response['ratings'][2]['value']) #metacritic score
    except:
        metacrit_score.append('null') #metacritic score

        #j += 1
        #found += 1
        #print(f"Processing Record {j} of {total} | {film}")
#     except:
#         print(f"Processing Record {j} of {total} | {film} not found - skipping.")
#         notfound += 1
        
#pct_movie_found = round(found/(notfound+found)*100,1)       
#print(f"{pct_movie_found}% of movies found")

# Coverting to a dataframe 


df = pd.DataFrame({'Title': titles, 'Metascore': metascore, 'Rated': rated,'Runtime': runtime,'Released': released, 'Year': year, 'Genre': genre,'Box Office': box_office,
                  'imdb_rating': imdb_rating, 'imdb_votes': imdb_votes,
                  'imdb_source': imdb_source, 'imdb_score': imdb_score,
                  'rotten_source': rotten_source, 'rotten_score': rotten_score,
                  'metacrit_source': metacrit_source, 'metacrit_score': metacrit_score})
df
#df = pd.DataFrame({'Title': titles, 'Metascore': metascore, 'Rated': rated,'Runtime': runtime,'Released': released, 'Year': year, 'Genre': genre,'Box Office': box_office,
                  #'imdb_rating': imdb_rating, 'imdb_votes': imdb_votes, 'imdb_source':imdb_source, 'imdb_score':imdb_score, 'rotten_source': rotten_source, 'rotten_score': rotten_score})
#df.to_csv("films.csv", encoding="utf-8", index=False)
```
